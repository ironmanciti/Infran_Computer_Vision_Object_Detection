```python
import sys
import torch
import matplotlib.pyplot as plt
plt.rcParams["savefig.bbox"] = 'tight'  # 그림 저장 시 여백을 최소화하도록 설정

from torchvision.transforms import v2  # torchvision의 transforms v2 API 불러오기
from torchvision.io import decode_image  # 이미지 디코딩 함수 (파일에서 Tensor로 변환)

torch.manual_seed(1)  # 랜덤 시드를 설정해 결과 재현성 보장

# Colab에서 실행 중이라면, 아래 URL에서 에셋(이미지)과 헬퍼 파일을 다운로드할 수 있습니다.
# https://github.com/pytorch/vision/tree/main/gallery/

# sys.path에 헬퍼 파일 경로 추가
sys.path.append("/content/vision/gallery/transforms")

from helpers import plot  # 헬퍼 함수 불러오기 (이미지 플롯을 위한 사용자 정의 함수)

# 이미지 파일 불러오기 및 디코딩
img = decode_image('/content/vision/gallery/assets/astronaut.jpg')
# '../assets' 폴더에 있는 'astronaut.jpg' 파일을 불러와 Tensor로 변환

# 이미지 정보 출력
print(f"{type(img) = }, {img.dtype = }, {img.shape = }")
```

## 기본 개념

**Torchvision Transforms**는 일반적인 **`torch.nn.Module`**처럼 동작합니다 (객체 생성 후 입력 데이터 전달 방식)  
- **Transform**을 생성합니다.  
- 입력 데이터를 전달합니다.  
- 변환된 출력을 얻습니다.



```python
# 이미지를 224x224 크기로 랜덤하게 잘라냄
transform = v2.RandomCrop(size=(224, 224))

# 변환 적용: 원본 이미지(img)를 랜덤 크롭
out = transform(img)

# 원본 이미지와 변환된 이미지 시각화
plot([img, out])
```

## 이미지 분류만 하고 싶은 경우  

이미지 분류에만 관심이 있다면, v2를 이용한 기본적인 분류 파이프라인은 다음과 같이 사용됩니다:


```python
# 이미지 분류용 변환 파이프라인 정의
transforms = v2.Compose([
    v2.RandomResizedCrop(size=(224, 224), antialias=True),  # 이미지를 224x224 크기로 랜덤 크롭 및 리사이즈
    v2.RandomHorizontalFlip(p=0.5),  # 50% 확률로 이미지를 좌우 반전
    v2.ToDtype(torch.float32, scale=True),  # 이미지를 float32로 변환하고 [0, 255] 값을 [0, 1]로 스케일링
    v2.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),  # 이미지 정규화 (이미지넷 평균 및 표준편차)
])

# 변환 적용: 원본 이미지(img)에 대해 정의한 변환을 순차적으로 적용
out = transforms(img)

# 원본 이미지와 변환된 이미지 시각화
plot([img, out])
```

## **탐지, 분할, 비디오 작업**

**`torchvision.transforms.v2`** 네임스페이스의 Torchvision 변환은 **이미지 분류**를 넘어서는 작업도 지원합니다:  
- **바운딩 박스 변환**  
- **세그멘테이션/탐지 마스크 변환**  
- **비디오 변환**

바운딩 박스를 사용하는 탐지 예제를 간단히 살펴보겠습니다.

```python
from torchvision import tv_tensors

# 바운딩 박스 정의
# BoundingBoxes 클래스는 바운딩 박스를 정의하고 변환 시 함께 처리할 수 있도록 지원합니다.
# format="XYXY": 바운딩 박스의 좌표 형식 (왼쪽 위 x, y, 오른쪽 아래 x, y)
# canvas_size=img.shape[-2:]: 이미지의 높이와 너비를 캔버스 크기로 설정
boxes = tv_tensors.BoundingBoxes(
    [
        [15, 10, 370, 510],   # 첫 번째 바운딩 박스 좌표
        [275, 340, 510, 510], # 두 번째 바운딩 박스 좌표
        [130, 345, 210, 425]  # 세 번째 바운딩 박스 좌표
    ],
    format="XYXY", canvas_size=img.shape[-2:]  # 이미지 크기에 맞게 캔버스 설정
)

# v2 변환 파이프라인 정의
transforms = v2.Compose([
    v2.RandomResizedCrop(size=(224, 224), antialias=True),  # 이미지를 224x224 크기로 랜덤 리사이즈 및 크롭
    v2.RandomPhotometricDistort(p=1),  # 이미지 색상 왜곡(밝기, 대비, 채도 등)을 적용
    v2.RandomHorizontalFlip(p=1),  # 100% 확률로 이미지를 좌우 반전
])

# 변환 적용
# img: 원본 이미지
# boxes: 원본 바운딩 박스
# out_img: 변환된 이미지
# out_boxes: 변환된 바운딩 박스
out_img, out_boxes = transforms(img, boxes)

# 변환 전후 바운딩 박스 타입 출력
print(type(boxes), type(out_boxes))

# 변환 전후 결과 시각화
# (img, boxes): 원본 이미지와 바운딩 박스
# (out_img, out_boxes): 변환된 이미지와 바운딩 박스
plot([(img, boxes), (out_img, out_boxes)])
```

위의 예제는 **객체 탐지(Object Detection)** 의 경우 입니다.  
만약 객체 분할(Object Segmentation) 또는 의미론적 분할(Semantic Segmentation)을 위한 **마스크**(`torchvision.tv_tensors.Mask`)나 **비디오**(`torchvision.tv_tensors.Video`)를 가지고 있다면, 변환(transform)에 똑같이 전달할 수 있습니다.

## TVTensors란 ?

**TVTensors**는 **`torch.Tensor`**의 하위 클래스입니다.  
사용 가능한 TVTensors에는 다음이 포함됩니다:
- **`torchvision.tv_tensors.Image`**: 이미지 데이터용.  
- **`torchvision.tv_tensors.BoundingBoxes`**: 바운딩 박스 데이터용.  
- **`torchvision.tv_tensors.Mask`**: 마스크 데이터용.  
- **`torchvision.tv_tensors.Video`**: 비디오 데이터용.
**TVTensors는 텐서**는 **`torch.Tensor`** 에서 지원하는 모든 작업(`.sum()` 같은 메서드나 `torch.*` 연산자 등)을 TVTensor에서도 동일하게 사용할 수 있습니다.

```python
# TVTensor로 이미지 생성
# 0~255 범위의 임의의 정수 값을 가지는 크기 (3, 256, 256)인 uint8 텐서를 생성
img_dp = tv_tensors.Image(torch.randint(0, 256, (3, 256, 256), dtype=torch.uint8))

# img_dp가 torch.Tensor의 인스턴스인지 확인
print(f"{isinstance(img_dp, torch.Tensor) = }")  # True 출력: TVTensors는 torch.Tensor의 하위 클래스

# img_dp의 dtype, shape, 전체 값의 합 출력
print(f"{img_dp.dtype = }")  # 데이터 타입: torch.uint8
print(f"{img_dp.shape = }")  # 텐서의 크기: (3, 256, 256) (3채널 RGB 이미지)
print(f"{img_dp.sum() = }")  # 텐서의 모든 값의 합 계산
```

**TVTensor 클래스**는 **변환(Transforms)** 의 핵심입니다. 변환은 주어진 입력을 처리하기 위해 먼저 객체의 **클래스**를 확인하고, 해당 클래스에 맞는 구현을 호출합니다.

아래는 동일한 객체 탐지용 변환을 사용하면서, 입력으로 **튜플 (이미지, 타겟 딕셔너리)**를 전달하고, 변환 결과에서도 **같은 구조**를 반환하는 예제입니다.

```python
print(boxes.shape)
print(boxes)
```

```python
# 타겟(target) 딕셔너리 정의
# 바운딩 박스와 레이블 정보를 포함하며, 변환 과정에서 무시될 추가적인 데이터도 포함
target = {
    "boxes": boxes,  # 바운딩 박스 정보 (tv_tensors.BoundingBoxes)
    "labels": torch.arange(boxes.shape[0]),  # 각 바운딩 박스에 대한 레이블 (0부터 시작하는 정수 배열)
}

# 위에서 정의한 변환(transforms) 재사용
# 이미지(img)와 타겟(target)을 함께 변환
out_img, out_target = transforms(img, target)

# 변환 전후 바운딩 박스 시각화
# (img, target["boxes"]): 원본 이미지와 바운딩 박스
# (out_img, out_target["boxes"]): 변환된 이미지와 바운딩 박스
plot([(img, target["boxes"]), (out_img, out_target["boxes"])])

print(out_target)
```
