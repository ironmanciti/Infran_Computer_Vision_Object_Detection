# 뇌 MRI를 위한 U-Net

**뇌 MRI에서 이상 부위를 분할하기 위해 사전 학습된 가중치를 사용한, 배치 정규화를 포함한 바이오메디컬 이미지 세그멘테이션용 U-Net 모델**

<img src="https://pytorch.org/assets/images/unet_brain_mri.png" alt="alt" width="50%"/>

```python
import torch

# PyTorch 허브에서 사전 학습된 U-Net 모델 로드
model = torch.hub.load('mateuszbuda/brain-segmentation-pytorch', 'unet',
    in_channels=3, out_channels=1, init_features=32, pretrained=True)
```

U-Net 모델은 [kaggle.com/mateuszbuda/lgg-mri-segmentation](https://www.kaggle.com/mateuszbuda/lgg-mri-segmentation)에서 제공되는 뇌 MRI 볼륨 데이터셋을 사용하여 이상 부위 분할을 위해 사전 학습된 모델입니다.  
사전 학습된 모델은 3개의 입력 채널, 1개의 출력 채널, 첫 번째 레이어에서 32개의 특징 필터를 요구합니다.

---

### 모델 설명

이 U-Net 모델은 다음과 같은 구조로 이루어져 있습니다:
- **4개의 블록 레벨**: 각 블록에는 두 개의 컨볼루션 레이어가 있으며, 배치 정규화(Batch Normalization)와 ReLU 활성화 함수가 포함되어 있습니다.  
  인코딩 부분에는 한 개의 맥스 풀링 레이어(Max Pooling Layer)가 포함되어 있으며, 디코딩 부분에는 대신 업컨볼루션 레이어(Up-convolutional Layer)가 사용됩니다.
- **컨볼루션 필터 수**: 각 블록에서 필터 개수는 32, 64, 128, 256이며, 병목 레이어(bottleneck layer)는 512개의 필터를 가집니다.
- **스킵 연결(Skip Connections)**: 인코딩 레이어의 출력이 디코딩 부분의 대응 레이어로 전달됩니다.

**입력**: 입력 이미지는 각각 사전 대비(pre-contrast), FLAIR, 사후 대비(post-contrast) 시퀀스의 3채널의 뇌 MRI 슬라이스를 한장의 컬러 이미지로 결합한  이미지 입니다.  
**출력**: 출력은 입력 이미지와 동일한 크기를 가진 이상 부위 확률 맵입니다(1채널).  
이를 통해 임계값 처리를 통해 이진 분할 마스크로 변환할 수 있습니다.


사전 학습된 모델을 위한 입력 이미지는:
- 3개의 채널을 가져야 하며,
- 256x256 픽셀로 크기 조정되어야 하고,
- 전체 3D 볼륨에 Z-score 정규화를 적용해야 합니다.

참고) 3채널이란,   
- 사전 대비(Pre-contrast) : 조영제를 주입하기 전에 촬영한 MRI 이미지  
- FLAIR (Fluid-Attenuated Inversion Recovery) : 뇌척수액(CSF)과 같은 체액 신호를 억제하여 병변(lesion)을 더 잘 드러내는 특수한 MRI 시퀀스  
- 사후 대비(Post-contrast) : 조영제를 주입한 후 촬영한 MRI 이미지

```python
import urllib

# 다운로드할 이미지의 URL과 저장할 파일 이름을 지정
url, filename = ("https://github.com/mateuszbuda/brain-segmentation-pytorch/raw/master/assets/TCGA_CS_4944.png", "TCGA_CS_4944.png")

# 파일 다운로드
urllib.request.urlretrieve(url, filename)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from torchvision import transforms

# 이미지 시각화 (원본 이미지)
input_image = Image.open(filename)

plt.figure(figsize=(6, 6))
plt.title("Downloaded Image (Original)")
plt.imshow(input_image)
plt.axis("off")
plt.show()
```

```python
# 입력 이미지를 전처리. ImageNet 정규화를 사용
preprocess = transforms.Compose([
    transforms.Resize((256, 256)),  # 크기 조정
    transforms.ToTensor(),  # 텐서로 변환
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # ImageNet 평균/표준편차 사용
])

# 이미지를 텐서로 변환하고 전처리 수행
input_tensor = preprocess(input_image)

# 모델 입력 배치를 만들기 위해 배치 차원 추가 (1차원 추가)
input_batch = input_tensor.unsqueeze(0)

# GPU가 사용 가능한 경우, 데이터를 GPU로 이동
if torch.cuda.is_available():
    input_batch = input_batch.to('cuda')  # 입력 배치를 GPU로 이동
    model = model.to('cuda')  # 모델을 GPU로 이동

# 모델 추론 (gradient 계산 비활성화)
with torch.no_grad():
    output = model(input_batch)  # 모델에 입력 배치를 전달해 예측 수행

print(output.shape)
```

```python
from PIL import Image
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Sigmoid를 적용해 확률 맵 생성
output_image = torch.sigmoid(output[0]).squeeze(0).cpu().numpy()
# 확률 맵을 이진화 (0 또는 1)
binary_mask = (output_image > 0.7).astype(np.uint8)

# 이진 마스크를 원본 이미지 크기로 조정
binary_mask_resized = Image.fromarray(binary_mask * 255).resize(input_image.size, Image.NEAREST)
binary_mask_resized = np.array(binary_mask_resized)

# 시각화
plt.figure(figsize=(12, 8))

# 원본 이미지
plt.subplot(1, 3, 1)
plt.title("Original Image")  # 제목: 원본 이미지
plt.imshow(input_image)  # 원본 이미지 표시
plt.axis("off")  # 축 제거

# 확률 맵
plt.subplot(1, 3, 2)
plt.title("Probability Map")  # 제목: 확률 맵
plt.imshow(output_image, cmap="viridis")  # 확률 맵 표시 (viridis 색상 맵 사용)
plt.colorbar()  # 색상 바 추가
plt.axis("off")  # 축 제거

# 이진 세그멘테이션 마스크 오버레이
plt.subplot(1, 3, 3)
plt.title("Segmentation Mask Overlay")  # 제목: 세그멘테이션 마스크 오버레이
plt.imshow(input_image, alpha=0.7)  # 원본 이미지 투명하게 표시
plt.imshow(binary_mask_resized, cmap=ListedColormap(['black', 'red']), alpha=0.3)  # 마스크를 빨간색으로 오버레이
plt.axis("off")  # 축 제거

# 전체 레이아웃 조정 및 시각화
plt.tight_layout()
plt.show()
```
