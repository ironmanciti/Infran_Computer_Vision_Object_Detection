Mask R-CNN 사전 학습된 모델을 **Penn-Fudan 데이터셋**에 대해 **파인튜닝(fine-tuning)**하는 방법을 설명합니다.
- COCO 데이터셋은 80개의 클래스 중 "person" 클래스가 포함되어 있지만, Penn-Fudan 데이터셋은 보행자 탐지 및 분할에 특화된 데이터셋입니다.
- 따라서 Penn-Fudan 데이터셋을 활용해 모델을 파인튜닝하면, 보행자 여부 분류 및 분할에 더 높은 정확도와 특화된 성능을 가진 모델로 최적화할 수 있습니다.

분할 마스크는 보행자의 각 인스턴스를 서로 다른 색상으로 구분합니다.

````python
import matplotlib.pyplot as plt
from torchvision.io import read_image  # 이미지를 읽어와 Tensor 형식으로 변환하는 함수

# 원본 이미지와 마스크 이미지 읽기
image = read_image("data/PennFudanPed/PNGImages/FudanPed00046.png")  # Penn-Fudan 이미지 경로
mask = read_image("data/PennFudanPed/PedMasks/FudanPed00046_mask.png")  # 해당 이미지의 세그멘테이션 마스크 경로

# 이미지와 마스크를 나란히 시각화
plt.figure(figsize=(16, 8))

plt.subplot(121)
plt.title("Image")
plt.imshow(image.permute(1, 2, 0))  # (C, H, W) 형태를 (H, W, C)로 변환해 시각화

plt.subplot(122)
plt.title("Mask")
plt.imshow(mask.permute(1, 2, 0))

plt.show()
````

각 이미지에는 해당하는 **세그멘테이션 마스크**가 있으며, 각 색상은 다른 인스턴스를 나타냅니다. 이제 이 데이터셋에 대해 `torch.utils.data.Dataset` 클래스를 작성해보겠습니다.

아래 코드에서 우리는 **이미지**, **바운딩 박스**, 그리고 **마스크**를 **`torchvision.tv_tensors.TVTensor`** 클래스로 감쌉니다. 이렇게 하면 **`torchvision`의 새로운 변환 API** ([Transforms API](https://pytorch.org/vision/stable/transforms.html))를 사용해 객체 탐지 및 세그멘테이션 작업에 필요한 변환을 적용할 수 있습니다.

구체적으로:
- **이미지 텐서**는 `torchvision.tv_tensors.Image` 클래스로 감싸집니다.  
- **바운딩 박스**는 `torchvision.tv_tensors.BoundingBoxes` 클래스로 감싸집니다.  
- **마스크**는 `torchvision.tv_tensors.Mask` 클래스로 감싸집니다.  

**`torchvision.tv_tensors.TVTensor`**는 **`torch.Tensor`**의 하위 클래스이기 때문에, 감싸진 객체도 기본적으로 **텐서**이며 **`torch.Tensor` API**를 그대로 상속받아 사용할 수 있습니다.

**`torchvision`의 `tv_tensors`**에 대한 더 많은 정보는 [이 문서](https://pytorch.org/vision/main/auto_examples/transforms/plot_transforms_getting_started.html#what-are-tvtensors)를 참조하세요.

Penn-Fudan 데이터셋을 위한 커스텀 데이터셋 작성

### 데이터셋 정의
객체 탐지, 인스턴스 세그멘테이션, 사람 키포인트 감지를 학습하기 위해 참조 스크립트는 새로운 커스텀 데이터셋을 쉽게 추가할 수 있도록 지원합니다.  
데이터셋은 torch.utils.data.Dataset 클래스를 상속받아 다음과 같은 두 가지 메서드를 구현해야 합니다:

1. `__len__`: 데이터셋의 총 이미지 개수를 반환.  
2. `__getitem__`: 데이터셋의 단일 샘플을 반환.


`__getitem__` 메서드는 다음 내용을 튜플로 반환해야 합니다:  

1. image:  
- torchvision.tv_tensors.Image 또는 순수 Tensor 형태로 [3, H, W] 크기를 가지거나,
- PIL Image 형식으로 (H, W) 크기의 이미지.  

2. target (딕셔너리):   
- boxes: torchvision.tv_tensors.BoundingBoxes 타입, 크기 [N, 4].
    - N은 바운딩 박스의 수이며, 좌표는 [x0, y0, x1, y1] 형식입니다.  
- labels: torch.Tensor 형태의 정수 값, 크기 [N].
    - 각 바운딩 박스의 레이블. 0은 항상 배경 클래스를 의미합니다.  
- image_id: 정수. 각 이미지의 고유 ID로 평가 시 사용됩니다.   
- area: torch.Tensor 타입의 실수 값, 크기 [N].  
    - 각 바운딩 박스의 면적이며, COCO 메트릭 평가 시 작은, 중간, 큰 박스 구분에 사용됩니다.  
- iscrowd: torch.Tensor 타입의 uint8, 크기 [N].
    - iscrowd=True인 인스턴스는 평가 시 무시됩니다.  
- (선택사항) masks: torchvision.tv_tensors.Mask, 크기 [N, H, W].
    - 각 객체의 세그멘테이션 마스크.

데이터 세트가 위의 요구 사항을 준수하는 경우 참조 스크립트의
교육 및 평가 코드 모두에 작동합니다.  
평가 코드는 `pycocotools`의 스크립트를 사용하며, `pip install pycocotools`로 설치할 수 있습니다.  

주의사항: labels 값  
- 모델은 클래스 0을 배경으로 처리합니다.  
- 데이터셋에 배경 클래스가 없는 경우, labels에 0을 포함하지 않아야 합니다.  
- 예를 들어, 클래스가 고양이와 개 두 가지뿐이라면:  
    - 고양이: 1  
    - 개: 2  
    - 이미지에 고양이와 개가 모두 있는 경우, labels는 [1, 2]가 됩니다.  

Aspect Ratio Grouping (선택사항)  
- 훈련 중 각 배치에 비슷한 비율의 이미지만 포함하려면 `get_height_and_width` 메서드를 구현하는 것이 좋습니다.  
- 이 메서드를 제공하지 않으면 `__getitem__`을 통해 모든 데이터를 불러오게 되어 속도가 느려질 수 있습니다.  

*** 아래 코드 중 iscrowd의 의미 ***  
iscrowd는 COCO 데이터셋에서 유래된 필드로, **객체가 "crowd"(군중)**인지 여부를 나타내는 플래그입니다.

COCO 데이터셋에서는 객체 탐지 및 분할 작업에서 객체가 너무 밀집되어(군중) 정확한 경계나 분할이 어려운 경우, 이 객체를 "crowd"로 표시합니다.  
- iscrowd=0: 해당 객체는 개별 객체로 처리됩니다.  
- iscrowd=1: 해당 객체는 "crowd"(군중)으로 간주됩니다.

````python
import os
import torch

from torchvision.io import read_image  # 이미지를 읽어와 텐서로 변환
from torchvision.ops.boxes import masks_to_boxes  # 마스크에서 바운딩 박스를 생성
from torchvision import tv_tensors  # TVTensors 클래스 (Image, BoundingBoxes, Mask 등)
from torchvision.transforms.v2 import functional as F  # 다양한 변환 기능 제공

# PennFudanDataset 클래스 정의
class PennFudanDataset(torch.utils.data.Dataset):
    def __init__(self, root, transforms):
        self.root = root  # 데이터셋의 루트 경로
        self.transforms = transforms  # 데이터에 적용할 변환 함수
        # PNG 이미지와 마스크 파일을 정렬하여 로드
        self.imgs = list(sorted(os.listdir(os.path.join(root, "PNGImages"))))
        self.masks = list(sorted(os.listdir(os.path.join(root, "PedMasks"))))

    def __getitem__(self, idx):
        # 특정 인덱스의 이미지 및 마스크 파일 경로 생성
        img_path = os.path.join(self.root, "PNGImages", self.imgs[idx])
        mask_path = os.path.join(self.root, "PedMasks", self.masks[idx])
        img = read_image(img_path)  # 이미지를 텐서로 읽어옴
        mask = read_image(mask_path)  # 마스크를 텐서로 읽어옴

        # 마스크에서 객체 ID를 추출 (객체마다 고유한 색상 값으로 인코딩)
        obj_ids = torch.unique(mask)
        # 첫 번째 ID는 배경(background)이므로 제거
        obj_ids = obj_ids[1:]
        num_objs = len(obj_ids)  # 객체 수 계산

        # 각 객체 ID에 대해 이진 마스크 생성
        masks = (mask == obj_ids[:, None, None]).to(dtype=torch.uint8)

        # 각 마스크에서 바운딩 박스 좌표 계산
        boxes = masks_to_boxes(masks)

        # 모든 객체는 동일한 클래스(사람)으로 설정
        labels = torch.ones((num_objs,), dtype=torch.int64)

        # 추가 메타데이터 정의
        image_id = idx  # 이미지 ID
        area = (boxes[:, 3] - boxes[:, 1]) * (boxes[:, 2] - boxes[:, 0])  # 바운딩 박스 면적 계산
        iscrowd = torch.zeros((num_objs,), dtype=torch.int64)  # 모든 객체는 crowd로 간주하지 않음

        # torchvision의 TVTensors로 이미지와 타겟 래핑
        img = tv_tensors.Image(img)

        target = {}
        target["boxes"] = tv_tensors.BoundingBoxes(boxes, format="XYXY", canvas_size=F.get_size(img))  # 바운딩 박스
        target["masks"] = tv_tensors.Mask(masks)  # 객체 마스크
        target["labels"] = labels  # 객체 클래스 라벨
        target["image_id"] = image_id  # 이미지 ID
        target["area"] = area  # 바운딩 박스 면적
        target["iscrowd"] = iscrowd  # crowd 여부

        # 변환 함수가 정의된 경우 변환 적용
        if self.transforms is not None:
            img, target = self.transforms(img, target)

        # 이미지와 타겟 반환
        return img, target

    def __len__(self):
        # 데이터셋 크기 반환 (이미지 개수)
        return len(self.imgs)
````

## **모델 정의**
이 튜토리얼에서는 [Mask R-CNN](https://arxiv.org/abs/1703.06870)을 사용할 것입니다.  Mask R-CNN은 [Faster R-CNN](https://arxiv.org/abs/1506.01497)을 기반으로 한 모델입니다.  
- **Faster R-CNN**: 이미지에서 잠재적인 객체의 **바운딩 박스**와 **클래스 점수**를 예측하는 모델입니다.

![Faster R-CNN 구조](https://pytorch.org/tutorials/_static/img/tv_tutorial/tv_image03.png)

- **Mask R-CNN**: Faster R-CNN에 **추가 분기(branch)**를 추가하여, 각 인스턴스에 대한 **세분화 마스크(Segmentation Mask)**도 예측합니다.

![Mask R-CNN 구조](https://pytorch.org/tutorials/_static/img/tv_tutorial/tv_image04.png)

### **TorchVision 모델을 수정하는 두 가지 주요 상황**

1. **사전 학습된 모델을 사용하고, 마지막 레이어만 미세 조정(Finetuning)하고 싶은 경우**  
2. **모델의 백본(Backbone)을 다른 것으로 교체하고 싶은 경우**  
   - 예: 예측 속도를 높이기 위해 백본을 변경.

### **1 - 사전 학습된 모델에서 미세 조정(Finetuning)**

COCO 데이터셋에서 **사전 학습된 모델(pre-trained model)**을 시작점으로 사용하고,  
특정 클래스에 맞게 미세 조정하고 싶다고 가정해 보겠습니다.  

*** FastRCNNPredictor란?  ***  

- Faster R-CNN은 **Region Proposal Network(RPN)**과 Fast R-CNN으로 구성된 객체 탐지(Object Detection) 모델입니다.  
- Fast R-CNN은 RPN이 제안한 영역(Region of Interest, ROI)을 입력으로 받아, 클래스 점수와 바운딩 박스 좌표를 예측하는 역할을 합니다.  
- FastRCNNPredictor는 분류기(classifier) 및 **박스 좌표 예측기(regressor)**를 사용자 정의할 수 있도록 제공되는 클래스입니다.

````python
import torchvision
from torchvision.models.detection.faster_rcnn import FastRCNNPredictor

# COCO 데이터셋에서 사전 학습된 Faster R-CNN 모델 불러오기
model = torchvision.models.detection.fasterrcnn_resnet50_fpn(weights="DEFAULT")

# 새로운 분류기를 추가하기 위해 기존 분류기(classifier) 교체
# 새로운 num_classes 정의: 1개의 클래스(person) + 배경(background)
num_classes = 2

# 분류기를 위한 입력 피처의 개수 가져오기
# Faster R-CNN의 ROI(Region of Interest) 헤드의 box_predictor에서 클래스 점수(cls_score)와 관련된 입력 피처 수
in_features = model.roi_heads.box_predictor.cls_score.in_features

# 사전 학습된 분류기를 새로운 분류기로 교체
# FastRCNNPredictor를 사용하여 사용자 정의 클래스 수(num_classes)에 맞게 분류기 생성
model.roi_heads.box_predictor = FastRCNNPredictor(in_features, num_classes)
````

### **2 - 모델 수정하여 다른 백본 추가하기**
기본 backbone인 ResNet-50 대신 ResNet-101이나 EfficientNet, MobileNet 등의 백본을 사용할 경우, 더 깊은 네트워크 구조나 효율적인 아키텍처를 통해 더 정교한 특징을 추출할 수 있습니다. 이로 인해 탐지 정확도와 세그멘테이션 성능이 개선될 수 있습니다.

ResNet이나 VGG와 같은 모델은 계산량이 많아 속도가 느릴 수 있습니다.
MobileNet과 같은 경량화된 네트워크는 계산량이 적어 추론 속도를 빠르게 만들 수 있습니다. 따라서 실시간 애플리케이션에 적합합니다.

````python
import torchvision
from torchvision.models.detection import FasterRCNN
# RPN에서 사용할 앵커(anchor) 생성 도구
# 다양한 sizes와 종횡비(aspect ratios)를 가지는 앵커를 공간적 위치에 따라 생성
from torchvision.models.detection.rpn import AnchorGenerator

# 사전 학습된 backbone 모델. 특징(feature)만 반환하도록 설정
backbone = torchvision.models.mobilenet_v2(weights="DEFAULT").features

# mobilenet_v2의 경우, 출력 채널 수는 1280이므로 명시적으로 설정
backbone.out_channels = 1280

# RPN설정: 공간 위치(spatial location)당 5개의 크기와 3개의 종횡비(aspect ratio)로 anchor 생성
# 각 특징 맵(feature map)은 서로 다른 크기와 비율을 가질 수 있으므로 Tuple[Tuple[int]] 형식을 사용
anchor_generator = AnchorGenerator(
    sizes=((32, 64, 128, 256, 512),),  # 앵커 크기
    aspect_ratios=((0.5, 1.0, 2.0),)  # 종횡비(너비 대비 높이 비율)
)

# Region of Interest(ROI) 풀링 설정:
# - 어떤 feature map을 사용할지 지정(featmap_names=['0'])
# - ROI를 rescale한 후 크기를 output_size=7로 설정
# - 샘플링 비율(sampling_ratio)은 2로 설정
roi_pooler = torchvision.ops.MultiScaleRoIAlign(
    featmap_names=['0'],  # 사용하려는 feature map 이름
    output_size=7,        # 풀링 후 ROI 크기
    sampling_ratio=2      # 샘플링 비율
)

# Faster R-CNN 모델 구성:
# - backbone: mobilenet_v2 백본 사용
# - num_classes: 클래스 수 (2 = 객체 클래스 1개 + 배경)
# - rpn_anchor_generator: RPN의 앵커 생성기 설정
# - box_roi_pool: ROI 풀링 설정
model = FasterRCNN(
    backbone,
    num_classes=2,  # 클래스 수 (1개의 객체 클래스 + 배경)
    rpn_anchor_generator=anchor_generator,  # RPN 앵커 생성기
    box_roi_pool=roi_pooler  # ROI 풀링
)
````

PennFudan 데이터셋용 객체 탐지 및 인스턴스 분할 모델  
======================================================================  

이번에는 사전 학습된 모델을 미세 조정(finetuning)하려고 합니다.  
데이터셋 크기가 매우 작기 때문에, **접근 방법 1**을 따를 것입니다.  

또한, **인스턴스 분할 마스크(instance segmentation masks)**도 계산하고자 하므로, **Mask R-CNN**을 사용할 것입니다.  

````python
import torchvision
from torchvision.models.detection.faster_rcnn import FastRCNNPredictor  # Faster R-CNN의 분류기 설정 도구
from torchvision.models.detection.mask_rcnn import MaskRCNNPredictor    # Mask R-CNN의 마스크 예측기 설정 도구

def get_model_instance_segmentation(num_classes):
    """
    Mask R-CNN 모델을 로드하고, 사용자 정의 클래스 수에 맞게 분류기와 마스크 예측기를 수정하여 반환.
    :param num_classes: 클래스 수 (객체 클래스 수 + 배경 클래스)
    :return: 사용자 정의 Mask R-CNN 모델
    """
    # 1. COCO 데이터셋에서 사전 학습된 Mask R-CNN 모델 로드
    model = torchvision.models.detection.maskrcnn_resnet50_fpn(weights="DEFAULT")

    # 2. 기존 분류기(classifier) 수정
    # - 기존 분류기의 입력 피처 수 가져오기
    in_features = model.roi_heads.box_predictor.cls_score.in_features
    # - FastRCNNPredictor를 사용하여 사용자 정의 클래스 수(num_classes)에 맞는 새로운 분류기 생성
    model.roi_heads.box_predictor = FastRCNNPredictor(in_features, num_classes)

    # 3. 마스크 예측기(mask predictor) 수정
    # - 기존 마스크 예측기의 입력 피처 수 가져오기
    in_features_mask = model.roi_heads.mask_predictor.conv5_mask.in_channels
    # - Mask R-CNN 내부에서 사용하는 히든 레이어 크기 설정 (256)
    hidden_layer = 256
    # - MaskRCNNPredictor를 사용하여 사용자 정의 클래스 수(num_classes)에 맞는 새로운 마스크 예측기 생성
    model.roi_heads.mask_predictor = MaskRCNNPredictor(
        in_features_mask,  # 입력 피처 수
        hidden_layer,      # 히든 레이어 크기
        num_classes        # 사용자 정의 클래스 수
    )

    return model
````

이것으로 `model`은 사용자 정의 데이터셋에서 **학습** 및 **평가**를 진행할 준비가 완료되었습니다.

이제 데이터 증강/변환을 위한 몇 가지 **도우미 함수(helper functions)**를 작성해 보겠습니다:

````python
from torchvision.transforms import v2 as T

def get_transform(train):
    """
    데이터 증강 및 변환 파이프라인을 생성하는 함수.
    :param train: 학습(train) 모드 여부를 나타내는 Boolean 값.
                  True일 경우 학습 데이터에 증강(transform)을 추가.
    :return: 데이터 변환을 처리하는 torchvision.transforms.Compose 객체.
    """
    transforms = []  # 변환을 담을 리스트

    if train:
        # 학습 모드일 경우, 랜덤 가로 뒤집기(Random Horizontal Flip) 추가
        # 50% 확률로 이미지를 가로로 뒤집음
        transforms.append(T.RandomHorizontalFlip(0.5))

    # 데이터를 float형으로 변환하고, 픽셀 값을 [0, 1] 범위로 정규화(scale=True)
    transforms.append(T.ToDtype(torch.float, scale=True))

    # 데이터를 순수 텐서(torch.Tensor)로 변환
    transforms.append(T.ToPureTensor())

    # 변환 리스트를 Compose로 묶어 반환
    return T.Compose(transforms)
````

`forward()` 메서드 테스트
=====================================  

데이터셋을 반복(iterate)하기 전에, 샘플 데이터를 사용해 모델이 **학습(training)**과 **추론(inference)** 시에 무엇을 기대할지 확인하는 것이 좋습니다.  


````python
import utils

# 1. 사전 학습된 Faster R-CNN 모델 로드 (ResNet-50 백본 + FPN 사용)
model = torchvision.models.detection.fasterrcnn_resnet50_fpn(weights="DEFAULT")

# 2. PennFudan 데이터셋 로드 및 변환 설정
# - 'data/PennFudanPed' 경로에 있는 데이터를 로드
# - 학습 모드에서 데이터 증강(transform) 적용
dataset = PennFudanDataset('data/PennFudanPed', get_transform(train=True))

# 3. 데이터 로더 생성
# - 배치 크기: 2
# - 데이터를 무작위로 섞음(shuffle=True)
# - collate_fn: 배치 데이터를 병합하는 데 사용하는 유틸리티 함수
data_loader = torch.utils.data.DataLoader(
    dataset,
    batch_size=2,         # 한 번에 2개의 샘플 로드
    shuffle=True,         # 데이터를 무작위로 섞음
    collate_fn=utils.collate_fn  # 배치 병합 함수
)

# 4. 학습 모드 테스트
images, targets = next(iter(data_loader))  # 데이터 로더에서 첫 번째 배치 가져오기
images = list(image for image in images)  # 이미지를 리스트 형식으로 변환
targets = [{k: v for k, v in t.items()} for t in targets]  # 타겟 데이터를 딕셔너리 리스트로 변환

# 모델의 forward() 호출 (학습 시)
# - 이미지는 입력 데이터
# - 타겟은 바운딩 박스, 라벨 등 객체 정보 포함
output = model(images, targets)  # 손실(losses)과 탐지 결과(detections) 반환
print(output)  # 손실 출력

# 5. 추론(inference) 모드 테스트
model.eval()  # 모델을 평가 모드로 전환
x = [torch.rand(3, 300, 400), torch.rand(3, 500, 400)]  # 임의의 입력 데이터 생성 (이미지 2개)
predictions = model(x)  # 추론 실행: 탐지 결과(predictions) 반환
print(predictions[0])  # 첫 번째 이미지의 탐지 결과 출력
````

이제 학습과 검증을 수행하는 **메인 함수**를 작성합니다:

````python
# 학습 및 평가를 위한 도우미 함수 불러오기
from engine import train_one_epoch, evaluate

device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')

# 데이터셋 클래스 수 정의 (배경 + 사람)
num_classes = 2

# PennFudan 데이터셋 로드 및 변환 정의
dataset = PennFudanDataset('data/PennFudanPed', get_transform(train=True))  # 학습용 데이터셋
dataset_test = PennFudanDataset('data/PennFudanPed', get_transform(train=False))  # 평가용 데이터셋

# 데이터셋을 학습용(train)과 테스트용(test)으로 분리
indices = torch.randperm(len(dataset)).tolist()  # 데이터셋 인덱스를 무작위로 섞음
dataset = torch.utils.data.Subset(dataset, indices[:-50])  # 마지막 50개를 제외한 부분을 학습용
dataset_test = torch.utils.data.Subset(dataset_test, indices[-50:])  # 마지막 50개를 테스트용

# 학습 및 검증 데이터 로더 정의
data_loader = torch.utils.data.DataLoader(
    dataset,
    batch_size=2,          # 배치 크기: 2
    shuffle=True,          # 데이터를 무작위로 섞음
    collate_fn=utils.collate_fn  # 배치 병합 함수
)

data_loader_test = torch.utils.data.DataLoader(
    dataset_test,
    batch_size=1,          # 배치 크기: 1
    shuffle=False,         # 데이터를 섞지 않음
    collate_fn=utils.collate_fn  # 배치 병합 함수
)

# 사용자 정의 Mask R-CNN 모델 생성
model = get_model_instance_segmentation(num_classes)

# 모델을 올바른 장치(GPU/CPU)로 이동
model.to(device)

# 옵티마이저(Optimizer) 설정
# - 학습 가능한 파라미터만 선택 (requires_grad=True)
params = [p for p in model.parameters() if p.requires_grad]
optimizer = torch.optim.SGD(
    params,                # 모델의 파라미터
    lr=0.005,              # 학습률
    momentum=0.9,          # 모멘텀
    weight_decay=0.0005    # 가중치 감소
)

# 학습률 스케줄러(Learning Rate Scheduler) 설정
lr_scheduler = torch.optim.lr_scheduler.StepLR(
    optimizer,             # 옵티마이저
    step_size=3,           # 3 에포크마다 학습률 감소
    gamma=0.1              # 학습률 감소 비율
)
````

````python
%%time
# 학습 반복 횟수 정의
num_epochs = 2

for epoch in range(num_epochs):
    # 한 에포크 동안 학습 (10번마다 진행 상황 출력)
    train_one_epoch(model, optimizer, data_loader, device, epoch, print_freq=10)
    # 학습률 스케줄러 업데이트
    lr_scheduler.step()
    # 테스트 데이터셋에서 평가 수행
    evaluate(model, data_loader_test, device=device)

print("That's it!")  # 학습 및 평가 완료 메시지 출력
````

따라서 한 에포크의 학습을 완료한 후, COCO 스타일의 mAP가 50 이상이고, 마스크 mAP가 65를 기록합니다.

데이터셋에서 하나의 이미지를 가져와 예측 결과를 확인해 봅니다.  

````python
import matplotlib.pyplot as plt
# bounding box와 segmentation mask를 이미지에 그리기 위한 유틸리티 함수
from torchvision.utils import draw_bounding_boxes, draw_segmentation_masks

# PennFudan 데이터셋에서 테스트할 이미지를 로드
image = read_image("data/PennFudanPed/PNGImages/FudanPed00046.png")

# 평가(evaluation) 시에 사용할 데이터 변환 정의
eval_transform = get_transform(train=False)

# 모델을 평가 모드로 전환
model.eval()

# 추론 실행 (torch.no_grad()로 그래디언트 비활성화)
with torch.no_grad():
    x = eval_transform(image)  # 이미지 변환 적용
    # RGBA 형식을 RGB로 변환하고, 디바이스(GPU 또는 CPU)로 이동
    x = x[:3, ...].to(device)
    predictions = model([x, ])  # 모델에 입력하여 예측 수행
    pred = predictions[0]  # 첫 번째 이미지의 예측 결과 가져오기

# 원본 이미지를 0~255 범위로 정규화하고 uint8 형식으로 변환
image = (255.0 * (image - image.min()) / (image.max() - image.min())).to(torch.uint8)
image = image[:3, ...]  # RGB 채널만 유지

# 예측된 라벨과 점수를 텍스트로 변환 (예: "pedestrian: 0.892")
pred_labels = [f"pedestrian: {score:.3f}" for label, score in zip(pred["labels"], pred["scores"])]

# 예측된 바운딩 박스 가져오기
pred_boxes = pred["boxes"].long()

# 바운딩 박스를 원본 이미지에 그리기 (빨간색)
output_image = draw_bounding_boxes(image, pred_boxes, pred_labels, colors="red")

# 예측된 마스크를 기준값(0.7) 이상인 픽셀만 남김
masks = (pred["masks"] > 0.7).squeeze(1)

# 원본 이미지에 분할 마스크 그리기 (파란색, 투명도 50%)
output_image = draw_segmentation_masks(output_image, masks, alpha=0.5, colors="blue")

# 최종 결과를 시각화
plt.figure(figsize=(12, 12))  # 이미지 크기 설정
plt.imshow(output_image.permute(1, 2, 0))  # 텐서를 (H, W, C) 형식으로 변환해 표시
````
